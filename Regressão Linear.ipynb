# Regressão Linear

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
dados = pd.DataFrame({'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50], 'C': [100, 200, 300, 400, 500]})
dados
```

```python
X_train, X_test, y_train, y_test = train_test_split(dados.drop(['C'], axis=1), dados.drop(['A', 'B'], axis=1), 
                                                    test_size=0.4, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

# Predições

```python
lm.predict([[2, 20]])
```
